```python
import xmltodict
import os
import pandas as pd
```

```python
def dic_to_lis(isr_data, fin_info):
    parse_data=[]
    for key,value in fin_info.items():
        if key =="Id":
            parse_data.append(value)
        elif key =='FullNm':
            parse_data.append(value)
        elif key =='ClssfctnTp':
            parse_data.append(value)
        elif key =='CmmdtyDerivInd':
            parse_data.append(value)
        elif key =='NtnlCcy':
            parse_data.append(value)
    else:
        parse_data.append(isr_data)
    
    return(parse_data)
```

```python
def data_to_csv(mine_info):
    rows=[]
    for i in range(len(mine_info)):
        try:
            isr_info = mine_info[i]['TermntdRcrd']['Issr']
            fin_value= mine_info[i]['TermntdRcrd']['FinInstrmGnlAttrbts']
            row= dic_to_lis(isr_data=isr_info, fin_info=fin_value)
            rows.append(row)

        except KeyError:
            try:
                isr_info=mine_info[i]['ModfdRcrd']['Issr']
                fin_value= mine_info[i]['ModfdRcrd']['FinInstrmGnlAttrbts']
                row= dic_to_lis(isr_data=isr_info, fin_info=fin_value)
                rows.append(row)

            except KeyError:
                isr_info=mine_info[i]['NewRcrd']['Issr']
                fin_value= mine_info[i]['NewRcrd']['FinInstrmGnlAttrbts']
                row= dic_to_lis(isr_data=isr_info, fin_info=fin_value)
                rows.append(row)
    return(rows)
```

```python
my_list = []
for files in os.listdir('task/'):
    xml_file=open(os.path.join('task/',files),'r',encoding='utf8').read()
    root = xmltodict.parse(xml_file)
    info=root['BizData']['Pyld']['Document']['FinInstrmRptgRefDataDltaRpt']['FinInstrm']
    data = data_to_csv(mine_info=info)
    
    col=['FinInstrmGnlAttrbts.Id','FinInstrmGnlAttrbts.FullNm',
      'FinInstrmGnlAttrbts.ClssfctnTp','FinInstrmGnlAttrbts.CmmdtyDerivInd',
     'FinInstrmGnlAttrbts.NtnlCcy','Issr']
    df = pd.DataFrame(data,columns=col)
    my_list.append(df)
```

```python
final_df = pd.concat(my_list,ignore_index=True)
```
